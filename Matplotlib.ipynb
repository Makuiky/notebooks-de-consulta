# Aspectos básicos generales

## Introducir datos según valores en los ejes

```python
fig, ax = plt.subplots() #genera la figura
ax.plot([1, 2, 3, 4], [1, 4.4, 2, 3]) #introduce los datos según coordenadas xy
plt.show() #necesario dependiendo del backend para mostrar la figura (en jupyter no es necesario)
```

## Partes de una figura

### Objeto figura

```python
fig =plt.figure() #figura sin ejes
```

```python
fig, ax = plt.subplots() #figura con dos ejes
plt.show()
```

### Ejes
No es lo mismo objeto ejes (axes) que eje (axis) para Matplotlib uno controla el espacio donde se va a generar el gráfico y el otro controla los ejes en si mismo

```python
fig, ax = plt.subplots() #figura con dos ejes
ax.set_title('hola')
ax.set_xlabel('horas')
ax.set_ylabel('envíos')
plt.show()
```

### Objeto Artist
Básicamente, todo lo visible en la figura es un objeto artist.

## Tipos de entrada

La entrada esperada por parte de matplotlib es un numpy.array. Clases similares a arrays como un objeto pandas o una matriz numpy pueden no tener el comportamiento esperado. Como convención es convertir los objetos en numpy.array antes de su visualización

```python
b = np.matrix([[1, 2], [3, 4]])
b_asarray = np.asarray(b)
b_asarray
```

La mayoría de los métodos también analizarán un objeto indexable por cadena como un dict, una matriz numpy estructurada o pandas.DataFrame. Matplotlib le permite proporcionar los datos de palabra clave y generar gráficos pasando las cadenas correspondientes a las variables xey.

# Figuras


## Crear figuras

```python
fig, ax = plt.subplots(figsize=(4,2)) 
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(4, 3), layout='constrained')
```

```python
fig, axs = plt.subplot_mosaic([['A', 'right'], ['B', 'right']],
                              figsize=(4, 3), layout='constrained')
for ax_name, ax in axs.items():
    ax.text(0.5, 0.5, ax_name, ha='center', va='center')
```

En este caso se generan subfiguras dentro de la figura inicial

```python
fig = plt.figure(layout='constrained', facecolor='lightskyblue')
fig.suptitle('Figure')
figL, figR = fig.subfigures(1, 2)
axL = figL.subplots(2, 1, sharex=True)
axR = figR.subplots(1,2)
```

# Ejes y subplots

# Crear y configurar ejes

```python
fig, axs = plt.subplots(1,2, figsize=(3.5, 2.5),
                        layout="constrained")
axs[0].plot([0,2],[1,1])
axs[1].plot([0.5,0.5],[0,3])
```

```python
fig = plt.figure()
axs=fig.add_axes([0, 0, 1, 1])
axs.plot([0,2],[1,1])
```

Para ver distintas opciones de visualización de datos visita el siguiente enlace: https://matplotlib.org/stable/plot_types/index.html

## Etiquetar ejes

```python
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.scatter(1,1, label='punto')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Distance [km]')
ax.set_title('Random walk example')
ax.legend()
```

## Límites y escalas de ejes

```python
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.set_yscale('log') #define escala del eje y (para eje x escribir set_xscale)
ax.set_xlim([20, 180]) #define el rango del eje x
```

## Modificar marcas de eje

```python
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.set_xticks([0,2,4,6])
ax.minorticks_off
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5))
ax.plot(np.arange(10))
ax.tick_params(top=True, labeltop=True, color='red', axis='x',
               labelcolor='green')
```

## Selección de relación de aspecto

```python
fig, axs = plt.subplots(ncols=2, figsize=(7, 2.5), layout='constrained')
axs[1].set_aspect(4)
```
